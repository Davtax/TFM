```python
import numpy as np
from general_functions import save_data
import matplotlib.pyplot as plt
```

```python
%%time
# Path the to file that we want to load
folder = 'data/'
file = 'FAQUAD_DQD_2HH_Systematic_Error'
extension = '.npy'
file_dic = folder + file + extension

data = np.load(file_dic, allow_pickle=True)  # Load the data, allow_pickle enable to load a list
print(data[-1])

results = data[0]
Gamma_vec = data[1]
tf_vec = data[2]

n_tf = len(tf_vec)
n_Gamma = len(Gamma_vec)
```

```python
density_matrix = []
probabilities = []
for temp in results:
    density_matrix.append(temp[0])
    probabilities.append(temp[1])
```

```python
final_result=[]
for i in range (0, n_tf*n_Gamma):
    temp=density_matrix[i][:,:,-1]
    final_result.append(temp)
```

```python
ET=1.17213075
u=2000
tau=4
l2 = tau * 0.4
l1 = l2 / 100

limit1 = 0.999
limit2 = 0.999

n_t=1e3

parameters={'ET':ET,'u':u,'tau':tau,'l1':l1,'l2':l2, 'limit1':limit1,'limit2':limit2, 'n_t':n_t}

save_data(file+'_reduced',[final_result,tf_vec,Gamma_vec,parameters,['final_density_matrix','tf_vec','Gamma_vec','parameters']])
```

```python
population_middle = np.zeros([n_tf, n_Gamma])
fidelity = np.zeros([n_tf, n_Gamma])
for i in range(0, n_tf):
	for j in range(0, n_Gamma):
		index = i * n_Gamma + j
		temp = probabilities[index]
		population_middle[i, j] = np.max(temp[:, 2])
		fidelity[i, j] = temp[-1, 1]
```

```python
fig, ax = plt.subplots()
pos = ax.imshow(fidelity.transpose(), origin='lower', cmap='jet', aspect='auto',
                extent=[tf_vec[0], tf_vec[-1], Gamma_vec[0], Gamma_vec[-1]], interpolation='quadric', vmin=0, vmax=1)
cbar = fig.colorbar(pos, ax=ax)
```
